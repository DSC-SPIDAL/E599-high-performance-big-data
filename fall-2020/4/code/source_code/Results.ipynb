```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
baseline = pd.read_csv('./baseline_train-clean-360.csv')
improved = pd.read_csv('./improved_train-clean-360.csv')
```

```python
plt.figure(figsize = (20,10))
plt.plot(list(range(1, 21)), baseline['avg_cer'], label = 'Baseline model')
plt.plot(list(range(1, 21)), improved['avg_cer'], label = 'Improved model')
plt.ylabel('Character Error Rate')
plt.xlabel('Epochs')
plt.xticks(ticks = list(range(1, 21, 2)))
plt.legend()
plt.savefig('cer.png')
```

```python
plt.figure(figsize = (20,10))
plt.plot(list(range(1, 21)), baseline['avg_wer'], label = 'Baseline model')
plt.plot(list(range(1, 21)), improved['avg_wer'], label = 'Improved model')
plt.ylabel('Word Error Rate')
plt.xlabel('Epochs')
plt.xticks(ticks = list(range(1, 21, 2)))
plt.legend()
plt.savefig('wer.png')
```
